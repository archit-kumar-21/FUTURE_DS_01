```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/raw/SuperstoreRaw.csv", encoding='latin1')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
#cleaning

df['Order Date']= pd.to_datetime(df['Order Date'])
df['Ship Date']= pd.to_datetime(df['Ship Date'])
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Year']= df['Order Date'].dt.year
```

```python
df['Month']= df['Order Date'].dt.month
```

```python
df['Month Name']= df['Order Date'].dt.month_name()
```

```python
#EDA

#####Revenue over time

sales_trend= df.groupby('Year')['Sales'].sum()

sales_trend.plot(kind='line', figsize=(10,5))
plt.title('Revenue Trend Over Time')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

```python
#######Top 10 selling products

top_products= df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='barh',figsize=(4,6))
plt.title('Top 10 Products by Revenue')
plt.show()
```

```python
#####Sales by Category

category_sales= df.groupby('Category')['Sales'].sum()

sns.barplot(x=category_sales.index,y=category_sales.values)
plt.title('Sales by Category')
plt.show()
```

```python
region_profit= df.groupby('Region')['Profit'].sum()

sns.barplot(x=region_profit.index, y=region_profit.values)
plt.title('Profit by Region')
plt.show()
```

```python
#######Category vs Profit

sns.boxplot(x='Category',y='Profit', data=df)
plt.title('Profit Distribution by Category')
plt.show()
```

```python
#discount vs Profit

sns.scatterplot(x='Discount',y='Profit',data=df)
plt.title('Discount Impact on profit')
plt.show()
```

```python
## Save cleaned data to folder 
df.to_csv('../data/cleaned/SuperstoreCleaned.csv',index=False)
```
